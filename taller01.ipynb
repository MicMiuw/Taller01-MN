#### 1. Graficar la diferencia de precios por token entre Longcontext y Shortcontext

```python
import numpy as np
```

```python
tokens = list(range(1_000_000))
```

```python
print(tokens[:10])
print(tokens[-5:])
```

```python
precio = []
for t in tokens:
    if t <= 272_000:
        precio.append((30 / 1_000_000) * t)
    else:
        costo_tramo1 = (30 / 1_000_000) * 272_000       
        costo_tramo2 = (45 / 1_000_000) * (t - 272_000)   
        precio.append(costo_tramo1 + costo_tramo2)
```

```python
import matplotlib.pyplot as plt
plt.plot(tokens, precio)
plt.xlabel('Tokens')
plt.ylabel('Precio')
plt.title('Precio vs Tokens')
plt.show()
```

### 2. Usando el método de la bisección, encuentre las raíces de la siguiente expresión cuadrática
$$f(x) = \frac{1}{4}(x^3 + 3x^2 - 6x - 8) = 0$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import bisect
```

```python
def f(x):
    return (1/4) * (x**3 + 3*x**2 - 6*x - 8)
```

```python
x = np.linspace(-5, 3, 1000)
y = f(x)

plt.figure(figsize=(9, 5))
plt.plot(x, y, 'b-', linewidth=2, label=r'$f(x) = \frac{1}{4}(x^3 + 3x^2 - 6x - 8)$')
plt.axhline(0, color='black', linewidth=1)  # eje x
plt.axvline(0, color='black', linewidth=1)  # eje y
plt.title('Gráfica de la función')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-4, 4)
plt.grid(True)
plt.legend()
plt.show()
```

```python
raiz1 = bisect(f, -4, -3)
raiz2 = bisect(f, -2, 0)
raiz3 = bisect(f, 1, 3)

print("Raíz 1:", raiz1)
print("Raíz 2:", raiz2)
print("Raíz 3:", raiz3)
```
